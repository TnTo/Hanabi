```python
memories = data['memories']
```

```python
end_memories = [m for m in memories if m.post.ended == True]
```

```python
ehints = [m.pre.hints for m in end_memories]
elifes = [m.post.lifes for m in end_memories]
eturn = [m.pre.turn for m in end_memories]
epoints = [m.post.score() for m in end_memories]
```

```python
actions = []
for m in memories:
    if m.action.hintColor or m.action.hintNumber:
        actions.append('H')
    elif m.action.played:
        actions.append('P')
    elif m.action.discarded:
        actions.append('D')
hints = [m.pre.hints for m in memories]
lifes = [m.post.lifes for m in memories]
turn = [m.pre.turn for m in memories]
points = [m.post.score() for m in memories]
```

```python
sns.countplot(hints, hue=actions)
```
